```python
import time
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

dir_path = os.getcwd()
data_dir = "data/"
```

These are the scores created by Gregor Samsa (@Octo_bery on twitter) from [this spreadsheet](https://docs.google.com/spreadsheets/d/1cSh4Jg3UU9g2FUqi98I-XmvloBU-MYH-KV1M2UcgWWg/edit#gid=0). 

```python
df_scores = pd.read_csv(data_dir + "scores.csv")
df_scores.columns = ["artists", "album", "score", "gender"]
df_scores.artists = df_scores.artists.str.lower()
df_scores.album = df_scores.album.str.lower()
df_scores.head()
```

```python
df_albums = pd.read_csv(data_dir + "albums.csv", sep = ";")
df_albums.head()
```

```python
df_id_scores = pd.merge(df_scores, df_albums, on = "album", how = "inner")
df_id_scores = df_id_scores.drop_duplicates()
df_id_scores = df_id_scores.iloc[:, :-1]
df_id_scores
```

```python
df_transcripts = pd.read_csv(data_dir + "full_transcripts.csv", sep = ";")
df_transcripts = pd.merge(df_id_scores, df_transcripts, on = "id")
df_transcripts = df_transcripts.drop_duplicates()
df_transcripts = df_transcripts.dropna()
df_transcripts
```

```python
df_transcripts.to_csv(data_dir + "album_transcripts.csv", index = False, sep = ";")
```
